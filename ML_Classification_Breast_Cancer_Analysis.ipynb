```python
# Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Importing Data
from sklearn.datasets import load_breast_cancer
cancer_data = load_breast_cancer()
```

```python
cancer_data.data.shape
```

```python
cancer_data.feature_names
```

```python
data = pd.DataFrame(cancer_data.data)
data.columns = cancer_data.feature_names 
data.head(10)
```

```python
data['diagnosis'] = cancer_data.target
data.head()
```

```python
data.describe()
```

```python
# Input Data
x = cancer_data.data

# Output Data
y = cancer_data.target

# splitting data to training and testing dataset.
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size =0.2,	random_state = 0)

print("xtrain shape : ", xtrain.shape)
print("xtest shape : ", xtest.shape)
print("ytrain shape : ", ytrain.shape)
print("ytest shape : ", ytest.shape)
```

```python
from sklearn.metrics import confusion_matrix

def accuracy(ytest, ypred):
  cm = confusion_matrix(ytest, ypred)
  c = cm[0, 0] + cm[1, 1]
  t = cm[0, 0] + cm[0, 1] + cm[1, 0] + cm[1, 1]
  print('Accuracy:',c/t * 100)
```

```python
# Fitting Logistic Regression on the input data
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0, max_iter=250)
classifier.fit(xtrain, ytrain)

# predicting the test set results
ypred = classifier.predict(xtest)
accuracy(ytest, ypred)
```

```python
#Using KNeighborsClassifier Method of neighbors class to use Nearest Neighbor algorithm
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 9, metric = 'minkowski', p = 2)
classifier.fit(xtrain, ytrain)

# predicting the test set results
ypred = classifier.predict(xtest)
accuracy(ytest, ypred)
```

```python
# Using SVC method of svm class to use Support Vector Machine Algorithm
from sklearn.svm import SVC
classifier = SVC(kernel = 'linear', random_state = 0)
classifier.fit(xtrain, ytrain)

# predicting the test set results
ypred = classifier.predict(xtest)
accuracy(ytest, ypred)
```

```python
# Using GaussianNB method of naïve_bayes class to use Naïve Bayes Algorithm
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(xtrain, ytrain)

# predicting the test set results
ypred = classifier.predict(xtest)
accuracy(ytest, ypred)
```

```python
# Using DecisionTreeClassifier of tree class to use Decision Tree Algorithm
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(xtrain, ytrain)

# predicting the test set results
ypred = classifier.predict(xtest)
accuracy(ytest, ypred)
```

```python
#Using RandomForestClassifier method of ensemble class to use Random Forest Classification algorithm
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 15, criterion = 'entropy', random_state = 0)
classifier.fit(xtrain, ytrain)

# predicting the test set results
ypred = classifier.predict(xtest)
accuracy(ytest, ypred)
```
